# 国际合作项目

中亚地区是“一带一路”倡议下“陆上丝绸之路”的重要桥头堡，也是全球重要的粮棉生产基地，拥有耕地5.7亿亩，人均耕地面积是我国的6倍。乌兹别克斯坦是中亚粮棉主产区，小麦和棉花产量分占中亚20%和64%。乌方农业机械化水平在中亚地区处于领先，但精量播种机、采棉机等农机装备不足20%，且老旧装备占比超过37%，严重制约农业生产。我国新疆与乌方同属寒旱粮棉产区，农业机械化水平领跑全国，小麦和棉花单产是乌方的3-4倍。乌方迫切希望与中方开展科技合作，借鉴新疆粮棉生产范式以实现产能提升和减贫增收。

鉴于乌兹别克斯坦与新疆在棉花品种、农艺措施以及种植模式方面存在显著差异，本项目聚焦于乌方农田存在的诸多农艺问题，如农田地块平整度欠佳、种植行距不统一（涵盖60、76、90厘米等多种行距），以及棉花产量预测方法精准度与时效性不足等技术问题，联合国内顶尖涉农高校及农机出口企业，聚焦采摘机构的仿形感知技术、数字化设计，以及智能产量监测技术的研究。通过运用“机械设计建模与动态仿真优化”相结合的方法，对不同行距的采摘机构进行优化设计，并研制快速换装装置。同时，探索低成本的农业信息多源感知技术以及多源数据深度融合方法，致力于攻克采棉机作业过程中的实时快速地面仿形与精准产量监测技术难题，实现农机与当地农艺的有效融合，进而推动国产采棉机本土适配技术的迭代升级。同时，项目填补了兵团在“一带一路”联合实验室建设的空白，项目联合乌兹别克斯坦顶尖涉农高校，建立棉花全程机械化示范基地，首次将“产学研用”一体化体系引入中亚，助力国产农业装备在乌兹别克斯坦的示范应用，具有显著的应用示范创新性。
