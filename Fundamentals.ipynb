```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含两个子图的图形对象，图形大小为9x4
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# 为了保证随机数的可重复性，设定随机数种子
np.random.seed(19680801)

# 生成一些随机测试数据，每组数据服从正态分布，标准差分别为6到9
all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

# 绘制小提琴图，显示中位数，不显示均值
axs[0].violinplot(all_data,
                  showmeans=False,
                  showmedians=True)
axs[0].set_title('Violin plot')

# 绘制箱线图
axs[1].boxplot(all_data)
axs[1].set_title('Box plot')

# 为每个子图添加水平网格线，并设置x轴刻度标签
for ax in axs:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(all_data))],
                  labels=['x1', 'x2', 'x3', 'x4'])
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')

# 显示图表
plt.show()
```

## 动力学建模及动力学分析

### 需要掌握的基础知识

- D'Alembert principle, 达朗贝尔原理
- Newton's laws of motion, 牛顿运动定律
- 拉格朗日方程
- 动力学模型本质是二阶常微分方程组
- 动力学模型的求解方法有数值法、微分方程法、积分方程法
    - 欧拉法
    - 龙格库塔法（参考Matlab官网教程）、(https://web.mit.edu/10.001/Web/Course_Notes/Differential_Equations_Notes/lec24.html)
    - 高斯-赛德尔方法
    - 打靶法
    - 有限差分法
    - 有限元法（Galerkin方法）
- 采用4-5阶Runge-Kutta方法求解常微分方程组时，注意刚性问题
- 通常采用无量纲化处理
- 二阶常微分方程组转换为状态方程（降维）

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义Lorenz方程
def lorenz(t, state, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# 初始条件
initial_state = [1, 1, 1]

# 参数
sigma = 10
rho = 28
beta = 8/3

# 时间范围
t_span = (0, 50)
t_eval = np.linspace(*t_span, 10000)

# 求解Lorenz方程
sol = solve_ivp(lorenz, t_span, initial_state, args=(sigma, rho, beta), t_eval=t_eval)

# 绘制结果
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2], lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor')
plt.show()
```
