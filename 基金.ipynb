# 国家自然科学基金辅导

## 基金申请那些事

优秀的申请书，关键在于**精炼**与**合乎逻辑**，用最简洁的语言传递信息，让评审专家快速抓住研究内容和创新点。

申请书的共性问题：

1. 整体创新性缺乏
  
  同质化、跟风现象严重：热衷随大流，整体创新能力不足

  创新性理解偏差：新技术或新方法的简单应用。

2. 关键科学问题凝练不足

  过大、过空。科学问题的表述冗长、堆砌术语，缺乏深度思考。

3. 立项依据与研究思路存缺陷

  立项依据不充分，表现在对研究必要性论证不充分，夸大研究意义。缺乏坚实的科学基础和合理的逻辑。

4. 研究方案缺乏独特性
   - 技术手段单一
   - 方案设计全面性不足
   - 方案简单堆砌
   - 盲目蹭热点




用戏剧思维抓住评审人的心

科学问题的凝练

直面真正的科学问题解决途径，独特的解决思路

对创新点的准确认知，研究内容不等同于创新点

符合科学逻辑，简洁明了的叙事方式

尽最大可能避免热门领域，如果你在做热门的课题，你可能已经落后了

尽最大可能避免同质化研究，保持独特性

避免项目组织上表象包装，具有宏大叙事风格特征的各类项目的立项依据保持清醒

对科学研究自身规范与价值观的尊重，坚持科学的体面与体面的科学

撰写态度上较真的精神

留白也是一种智慧

申请书的核心三要素

1. 做什么？
2. 为什么做？
3. 怎样做？

1. 选择一个独特的、有趣或重要的科学问题，而不是创造一个问题
2. 阐述为什么要针对该问题开展研究（科学意义，重要性、前人研究进展与研究方法的局限性以及符合科学逻辑推演的本申请的研究内容）
3. 提出一个独特的研究思路、策略或方案以及可能的符合科学逻辑Reasoning
